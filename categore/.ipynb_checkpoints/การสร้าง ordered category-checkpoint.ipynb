```python
from pandas.api.types import CategoricalDtype
import pandas as pd
```

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/staff.csv', index_col='empID')
df
```

```python
df.position.unique()
```

```python
df.sort_values('position')
```

```python
df['position_cat'] = df['position'].astype( CategoricalDtype([ 'ผู้จัดการ', 'รองผู้จัดการ', 'ผู้ช่วยผู้จัดการ','ผู้ชำนาญงาน', 'พนักงานระดับต้น' ]
                                                                        , ordered=True))
                                       

df.sort_values('position_cat')
```
