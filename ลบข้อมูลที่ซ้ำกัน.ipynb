```python
import pandas as pd
```

```python
url='https://github.com/prasertcbs/tutorial/raw/master/duplicates.csv'
df=pd.read_csv(url)
df
```

```python
df.drop_duplicates()     #duplicates หมายความว่าทุกเเถวจะต้องเหมือนกัน ถึงจะ drop
```

```python
df.drop_duplicates(subset=('manufacturer'))    #subset คือสนใจตัวที่ซ้ำในcolที่สนใจเเล้ว drop      inplace= true คือเเทนข้อมูลเก่า
```

```python
df.drop_duplicates(subset= ['manufacturer' , 'model' ] ) 
```
