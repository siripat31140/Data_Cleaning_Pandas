```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.head()
```

```python
df.pivot_table(index='class' , columns='drv' , values= 'model' ,  aggfunc='count' )   
#ถ้าพิมพ์แบบนี้จะ eror เพราะ valueจะรับเเค่ตัวเลย เเต่model เป็น str เราจะไปกำหนดaggfunc='mean' ไม่ได้ต้องใช้ counts
```

```python
df.pivot_table(index='class' , columns='drv' , values= 'model' ,  aggfunc='count' ,fill_value=0 , margins= True , margins_name='Grand total' )   
#เเทนค่า nan = o fill_value=0
# margins= True เเสดงผลรวม
#margins_name='Grand total ตั้งชื่อ
```

```python
df.pivot_table(index='class' , columns='drv' , values= 'model' ,  aggfunc='count' ,fill_value=0 , margins= True , margins_name='Grand total' ).astype('int')
#astype('int') ทำให้เป็นจำนวนเต็ม
```

```python
df.pivot_table(index='class' , columns='drv' , values= 'model' ,  aggfunc='count' ,fill_value=0 , margins= True , margins_name='Grand total' ).astype('int').plot(kind = 'barh');
```

```python
df.pivot_table(index='class' , columns='drv' , values= 'model' ,  aggfunc='count' ,fill_value=0  ).astype('int').plot(kind = 'barh');
```
