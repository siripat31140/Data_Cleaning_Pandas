## Example 1: remove currency code 

```python
df=pd.read_table('https://github.com/prasertcbs/tutorial/raw/master/currency.tsv', index_col='Year')
df
```

```python
df.TV.str.replace('\D', '').astype('int')  #ลบทุกอย่างยกเว้นตัวเลข เเล้วเเปลงเป็น int
```

```python
import re    #สร้าง function
def remove_nondigit(v):
    return int(re.sub(r'\D', '', v))
```

```python
remove_nondigit('usd123,000')
```

```python
df.applymap(remove_nondigit)  #ใช้ function all columns
```

```python
df2=df.applymap(remove_nondigit)
df2
```

```python
df2.plot()
```

```python
df.applymap(lambda v: int(re.sub(r'\D', '', v)))       #เอา lambda มาใช้ใน appltmap
```
