## Example 2: mpg

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df[:10]
```

```python
df.manufacturer.apply(str.upper) #ทำให้ตัวใหญ่
```

```python
df.manufacturer.map(str.upper)
```

```python
df[['manufacturer', 'model', 'class']].apply(lambda s: s.str.upper())  #หลายๆ col ก็ได้
```

```python
df.apply(lambda d: f'{d.manufacturer} ({d.model} {d.trans})', axis=1)  #หรือรวมข้อความกันก็ได้
```

```python
df
```

```python
df['kml'] = df.apply(lambda d: f'{d.cty * 0.425143707:4.1f}/{d.hwy * 0.425143707:4.1f}', axis=1)
df
```

```python
df.mean()
```

```python
df.select_dtypes(include=[np.number]).drop('year', axis=1)           #เลือกชนิดข้อมูลที่เป็นตัวเลขลบyearในcol
```

```python
df.select_dtypes(include=[np.number]).drop('year', axis=1).apply(stats.hmean)  #ไปหา hmean
```
