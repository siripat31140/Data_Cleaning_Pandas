## ตอนที่2    อยากให้ในเเต่ละเเถว มี Nanกี่ตัว

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/qnaire_missing_values.csv',
              index_col='obs')
df
```

```python
df.dropna(thresh=10)  #เลือกมาก10 col ที่ไม่มีค่า Nan 
```

```python
df.columns
```

```python
df.dropna(thresh=8 , subset=[ 'age', 'eq1', 'eq2', 'eq3', 'eq4', 'eq5', 'eq6', 'eq7', 'eq8',
       'eq9', 'eq10'])  #สนใจเเค่เเต่ในsubset  ส่วนgenderไม่ต้องสนว่าจะมี Nan 
```

```python
df_dd = df.dropna(thresh=10 , subset=[ 'age', 'eq1', 'eq2', 'eq3', 'eq4', 'eq5', 'eq6', 'eq7', 'eq8',
       'eq9', 'eq10']) 
df_dd
```
