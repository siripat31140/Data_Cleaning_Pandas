```python
import pandas as pd
```

```python
df=pd.read_table('https://github.com/prasertcbs/tutorial/raw/master/qnaire_fill_missing.tsv',
                index_col='obs', parse_dates=['dob'])
df
```

```python
df.fillna(df.mean()['csi1' : 'q3'] ,  inplace=True)
df
```

```python
df.describe()
```

```python
df.own_car.value_counts()
```

```python
df.own_car.fillna('False' ,inplace = True)
df
```

```python
df.own_house.value_counts()
df.own_house.fillna('False' ,inplace = True)
df.salary.fillna(df.mean()['salary'] , inplace = True)
df
```

```python
df.dob.fillna(pd.to_datetime('2001-01-16') , inplace = True)   #ถ้าไม่ pd.to_datetime มันจะกลายเป็น str
df
```

```python
df.gender.value_counts()
```

```python
df.gender.fillna('F' ,inplace = True)
df
```
