# use regex to clean data

https://en.wikipedia.org/wiki/List_of_unicorn_startup_companies

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
print(f'pandas version: {pd.__version__}')
```

```python
from datetime import datetime
str(datetime.now())
```

```python
dfs=pd.read_html('https://en.wikipedia.org/wiki/List_of_unicorn_startup_companies',
                 match='Valuation', header=0)
len(dfs)
```

```python
df=dfs[0]
df.head()
```

```python
df.info()
```

```python
df[df.Valuation.str.contains('\[')]
```

```python
df['Valuation'].str.replace('\[.*\]\.', '')
```

```python
df['aa']=df['Valuation'].str.replace('\[.*\]\.', '')
df
```

```python
df['bb']=df['aa'].str.replace('[^\d\.]', '')
df
```

```python
df['cc']=df['Valuation'].str.replace('(\[.*\]\.|[^\d\.])', '').astype('float')
```

```python
df
```

```python
df.info()
```

```python
df['dd']=pd.to_datetime(df['Valuation date'].str.replace('\[.*\]', ''))
df
```

```python
df['Valuation']=df['Valuation'].str.replace('(\[.*\]\.|[^\d\.])', '').astype('float')
df['Valuation date']=pd.to_datetime(df['Valuation date'].str.replace('\[.*\]', ''))
```

```python
df
```

```python
df.drop(columns=['aa', 'bb', 'cc', 'dd'], inplace=True)
df
```

```python
df.groupby('Country').describe()
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,5))
df.Country.value_counts().sort_values(ascending=True).plot(kind='barh', color='.75', width=.8);
```

```python
plt.figure(figsize=(6,5))
df.groupby('Country')['Valuation'].sum().sort_values().plot(kind='barh', color='.75', width=.8)
```
