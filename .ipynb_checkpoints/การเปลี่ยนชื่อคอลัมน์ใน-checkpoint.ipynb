```python
import pandas as pd
```

```python
df = pd.read_table('https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/BaskinRobbins.tsv')
df.head()
```

```python
 df.rename(columns={ 'Flavour' : 'Flav'} ,  inplace=True)   #เป็นชื่อcol เเล้ว inplacegเเทนที่ในข้อมูลเก่า
df.head()
```

```python
df.rename(columns={ 'Flavour' : 'Flav' , 'Calories': 'cal'} ,  inplace=True)
df.head()
```

```python
df.columns = df.columns.str.replace(' ' , '_' )   #replace พื้นที่ว่าเเทนด้วย_
df.head()
```

```python
df.columns = df.columns.str.lower(  )   #str.lower ทำใ้ตัวอีกษรตัวเล็กลง
df.head()
```

#### สามารถรวมสองคำสั่งได้

```python
df.columns = df.columns.str.replace(' ' , '_' ).str.lower()
df.head()
```
